```python
import pandas as pd

## 파일 읽기
total_temp = pd.read_csv('./data/weather_data_201501_202412.csv',encoding='CP949')
total_temp.info()
total_temp.describe()
```

```python
total_temp.isna().sum()
```

```python
total_temp['최심적설(cm)'] = total_temp['최심적설(cm)'].fillna(0)
total_temp = total_temp.dropna()
print( total_temp.isna().sum())

# 결측치 지역-월을 찾고 평균으로 na값 채움
# 연도 / 월 쪼개기 = 나중에 보여주는 목적에 따라 groupby
total_temp['years'] = total_temp['일시'].str.split('-').str[0].astype(int)
total_temp['month'] = total_temp['일시'].str.split('-').str[1].astype(int)
```

```python
# 전체 강원도 지역 중 기상청에서 수집한 강원도 기상 측정 지역 확인 (9개)
gangwon= ['춘천', '원주', '철원', '화천', '양구', '인제', '홍천', '횡성', '영월', '평창', '정선', '고성', '속초', '양양', '강릉', '동해', '삼척']
gangwon_list = []
for i in total_temp['지점명'].drop_duplicates(): 
    if i in gangwon:
        gangwon_list.append(i)
gangwon_list
```

```python
# 강원도 지역의 속한 데이터프레임 뽑기 
gangwon_keyword = "|".join(gangwon_list)
gangwon_total = total_temp[total_temp['지점명'].str.contains(gangwon_keyword)].reset_index(drop=True)
gangwon_total
gangwon_total.to_csv('./data/preprocessing/gangwon_temp.csv')

```
