```python
dfs = {
    '주택용': chung_house,
    '일반용': chung_general,
    '교육용': chung_edu,
    '산업용': chung_industry,
    '농사용': chung_farm,
    '심야': chung_night,
    '가로등': chung_light,
    '합계' : chung_total
}

all_frame = []
for name, df in dfs.items():
    temp = df.copy()
    temp['계약종별'] = name
    all_frame.append(temp)

gangwon_type = pd.concat(all_frame,ignore_index=True)
gangwon_type.groupby(['years'])['전력량'].mean()
sns.boxplot(data=gangwon_type, x='계약종별', y='전력량')
plt.title('충청도 계약종별 전력량',fontweight='bold', fontsize=16)
print(gangwon_type.keys())
dfs.keys()
```
