```python
import pandas as pd

## 파일 읽기
total_temp = pd.read_csv('C:/skn24/2601_mini_team/EDA_MINI_1TEAM/jaehoon/미니프로젝트/data/weather_data_201501_202412.csv',encoding='CP949')
total_temp.info()
total_temp.describe()
```

```python
total_temp.isna().sum()
```

```python
total_temp['최심적설(cm)'] = total_temp['최심적설(cm)'].fillna(0)
total_temp = total_temp.dropna()
print( total_temp.isna().sum())

# 결측치 지역-월을 찾고 평균으로 na값 채움
# 연도 / 월 쪼개기 = 나중에 보여주는 목적에 따라 groupby
total_temp['years'] = total_temp['일시'].str.split('-').str[0].astype(int)
total_temp['month'] = total_temp['일시'].str.split('-').str[1].astype(int)
total_temp['HDD'] = (18 - total_temp['평균기온(°C)']).clip(lower=0)
total_temp['CDD'] = (total_temp['평균기온(°C)'] - 24).clip(lower=0)
```

```python
total_temp['지점명'].unique()
```

```python
chung = [
    "천안", "아산", "서산", "당진", "공주", "보령", "논산", "계룡",
    "홍성", "예산", "태안", "금산", "부여", "서천", "청양", "청주", "충주", "제천",
    "보은", "옥천", "영동", "괴산", "음성", "단양", "증평", "진천","대전", '세종'
]



```

```python
# 전체 강원도 지역 중 기상청에서 수집한 강원도 기상 측정 지역 확인 (9개)
chung_list = []
for i in total_temp['지점명'].drop_duplicates(): 
    if i in chung:
        chung_list.append(i)
chung_list
```

```python
# 강원도 지역의 속한 데이터프레임 뽑기 
chung_keyword = "|".join(chung_list)
chung_total = total_temp[total_temp['지점명'].str.contains(chung_keyword)].reset_index(drop=True)
chung_total
chung_total.to_csv('C:/skn24/2601_mini_team/EDA_MINI_1TEAM/jaehoon/미니프로젝트/data/preprocessing/충청도 데이터/chung_temp.csv')

```
