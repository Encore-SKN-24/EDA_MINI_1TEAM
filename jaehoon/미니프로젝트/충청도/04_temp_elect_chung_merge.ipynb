```python
import pandas as pd
import glob

file_path = 'C:/skn24/2601_mini_team/EDA_MINI_1TEAM/jaehoon/미니프로젝트/data/preprocessing/충청도 데이터/*.csv'
file_glob = glob.glob(file_path)

contract_list = [pd.read_csv(file,encoding='CP949') for file in file_glob]
chung_temp = pd.read_csv('C:/skn24/2601_mini_team/EDA_MINI_1TEAM/jaehoon/미니프로젝트/data/preprocessing/충청도 데이터/온도/chung_temp.csv',encoding='utf-8')
chung_temp['지점명'].unique()
```

```python
# 계약종별 날씨, 전력 데이터 merge 저장
for i in contract_list:
    chung_merge = pd.merge(chung_temp,i,left_on=['years','지점명','month'], right_on = ['years','지점명','month'],how='inner')
    chung_merge = chung_merge.drop_duplicates()
    #결측치 포함 열 제거 
    chung_merge = chung_merge.dropna(how='any',axis=1)
    file_name = chung_merge['계약종별'].unique()[0]
    chung_merge.to_csv(f'C:/skn24/2601_mini_team/EDA_MINI_1TEAM/jaehoon/미니프로젝트/data/preprocessing/충청도 데이터/머지/{file_name}_elect_temp_gangwon.csv')
```
