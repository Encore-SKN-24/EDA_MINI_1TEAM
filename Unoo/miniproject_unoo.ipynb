```python
# 기타 결측치 '해당지역'의 '해당 월' 평균으로 채우기 

month_list = list(range(1,13))
region_list = weather_df['지점명'].unique()
# print(region_list)

for column in weather_df.iloc[:,3:].columns:
    # 1. 원본 결측치 위치를 기억
    nan_mask = weather_df[column].isna()
    # nan_mask_F_count = weather_df[column].isna().sum().item()
    # nan_mask_F_count
    

    # 2. 월 결측치 위치를 기억
    for m in month_list:
        month_mask = (weather_df['일시'].dt.month == m)

        # 3. 해당 지역 위치 기억
        for region in region_list:
            region_mask = weather_df['지점명'] == region
        
            region_month_avg = weather_df.loc[month_mask & ~nan_mask & region_mask, column].mean()

            weather_df.loc[month_mask & nan_mask & region_mask, column] = region_month_avg


```

```python
# 내 코드 검증 방식 '해당 지역' '해당 월' 둘다 비교
# 실제 '해당지역' '해당 월'로 데이터 잘 들어갔는지 확인
# 예: '서울' 지역의 데이터 중 원래 결측치였을 법한 구간 확인
sample_region = '진도군'
sample_month = 6
print('서울' in region_list)

region_month_condition = (weather_df['지점명'] == sample_region) & (weather_df['일시'].dt.month == sample_month)
print(weather_df[region_month_condition])
```

### 전력소모데이터 확인

```python
# 전력 소모량 데이터 확인

elect_usage_df = pd.read_csv('./data/elect_2015_2024.csv', encoding='cp949')

# 데이터 확인 -> 결측치 및 이상치 없음 확인
elect_usage_df.info()
print(elect_usage_df['계약종별'].unique())
print()
print(elect_usage_df.columns)
print(list(elect_usage_df.columns))
print("Index name is ", elect_usage_df.index.names)
display(elect_usage_df.describe())


# 'Unamed: 0' 열 drop
elect_usage_df = elect_usage_df.drop('Unnamed: 0', axis=1)
display(elect_usage_df)
```

```python
# DataFrame 알맞게 변환
# 연도와 월을 행으로 넣고 datetime 형식으로 바꾸기

elect_usage_df = elect_usage_df.melt(id_vars=['연도', '시도', '시군구', '계약종별'], 
                  var_name='월', 
                  value_name='지역과 계약종별별 전력사용량')

# '월' 컬럼에서 '월' 글자 제거, int 변환
elect_usage_df['월'] = elect_usage_df['월'].str.replace('월', '').astype(int)

# 연도와 월을 합쳐서 '일시' 컬럼 만들기 (날짜 형식)
# 분석을 위해 해당 월의 1일로 설정
elect_usage_df['일시'] = pd.to_datetime(elect_usage_df['연도'].astype(str) + '-' + elect_usage_df['월'].astype(str) + '-01')

# 정렬
elect_usage_df = elect_usage_df.sort_values(by=['일시', '시군구', '계약종별']).reset_index(drop=True)

elect_usage_df['일시'] = pd.to_datetime(pd.DataFrame({
    'year': elect_usage_df['연도'],
    'month': elect_usage_df['월'],
    'day': 1    # 실제로 일은 없지만 1로 통일
}))

# 결과 확인
display(elect_usage_df[['연도', '월', '일시']].head())
print(elect_usage_df['일시'].dtype) # datetime64[ns] 인지 확인


display(elect_usage_df)
```

```python
# 경상 지역별로 dataframe 분리
gsn_elect_df = elect_usage_df[elect_usage_df['시도'] == '경상북도']
gss_elect_df = elect_usage_df[elect_usage_df['시도'] == '경상남도']
busan_elect_df = elect_usage_df[elect_usage_df['시도'] == '부산광역시']
ulsan_elect_df = elect_usage_df[elect_usage_df['시도'] == '울산광역시']

# unique 값 확인
print(elect_usage_df['계약종별'].unique())
print(elect_usage_df['시도'].unique())
print(ulsan_elect_df['시군구'].unique())
print(busan_elect_df['계약종별'].unique())

# 경상 지역별 데이터 행 개수 확인 / 이유는 계약종별 가로등부터 합 계 까지 8개
print(len(gsn_elect_df)/8)
```

```python
# 결론은 지역마다 차이 큼!
# 지역 & 계약종별 확인! -> 서울은 종로구는 일반용이 가장 높게 나오고 
# 경상은 산업용이 좀 더 높게 나오는데 또 부산은 일반용과 주택용도 높게 나옴


# '합계'는 너무 커서 제외
df_plot = elect_usage_df[elect_usage_df['계약종별'] != '합 계'].copy()
df_plot = df_plot[df_plot['시도'] == '부산광역시']

# 그래프 생성
plt.figure(figsize=(14, 7))

# 계약종별마다 반복문을 돌며 선을 하나씩 추가 (꺾은선)
for kind in df_plot['계약종별'].unique():
    # 특정 계약종별만 추출 후 날짜순 정렬
    temp_df = df_plot[df_plot['계약종별'] == kind].sort_values('일시')
    

    # 핵심: '일시'별로 모든 시군구의 평균을 구함
    # 이렇게 해야 지그재그가 사라지고 매끈한 선이 나옵니다.
    avg_df = temp_df.groupby('일시')['지역과 계약종별별 전력사용량'].mean().reset_index()
    
    # 정렬 확인 후 출력
    avg_df = avg_df.sort_values('일시')
    plt.plot(avg_df['일시'], avg_df['지역과 계약종별별 전력사용량'], 
             label=kind, marker='o', markersize=3, linewidth=2)



    # '시군구' 로만 보고 싶으면 위에건 주석하고 아래걸로 plot
    # plt.plot(temp_df['일시'], temp_df['지역과 계약종별별 전력사용량'], 
    #          label=kind, 
    #          linewidth=2,     # 선 굵기
    #          marker='o',      # 꺾이는 지점에 점 표시 (선택사항)
    #          markersize=4)


# 범례 바깥쪽 배치
plt.legend(title='계약종별', bbox_to_anchor=(1.02, 1), loc='upper left')

# 레이아웃 자동 조정
plt.tight_layout()

# 그래프 출력
plt.show()
```
